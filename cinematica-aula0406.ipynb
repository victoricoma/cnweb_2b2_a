# Fórmula da velocidade
fonte: [https://www.todamateria.com.br/formulas-de-cinematica/]

```
def calculaVelMedia(deltaS, deltaT):
  velMedia = (deltaS/deltaT)
  return velMedia
```

## Teste Serviço Nuvem: Velocidade Média

```
print("Vm=", calculaVelMedia(30, 15))
```

# Função horária do deslocamento

```
def calculaDesloc(inicialS, deltaT, v):
  d = inicialS + (v*deltaT)
  return d
```

## Testes de Função : Horária de Deslocamento

```
print("Deslocamento= ",calculaDesloc(0, 15, 30))
```

# Equação de Torricelli

```
import math

def calTorricelli(vZero, acel, des):
  t=2*acel*des
  v=vZero+t
  res = round(math.sqrt(v), 2)
  return res
```

## Testes de Torricelli

```
print("Calculo de Torricelli: ",calTorricelli(0,3,8))
```
